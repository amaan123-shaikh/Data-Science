```python
Light=input("light:").lower()
if Light=="red":
    print("STOP")
elif Light=="yellow":
    print("LOOK")
elif Light=="green":
    print("GO")
else:
    print("LIGHT IS BROKEN")
```

```python
x="hellow world"
count=0
vowels="aeiou"
for char in x:
    if char in vowels:
        count+=1
print(count)
```

```python
x="hellow world"
count=""
vowels="aeiou"
for char in x:
    if char in vowels:
        count+=char
print(count)
```

```python
x="hellow world"
count=""
vowels="aeiou"
for char in x:
    if char in vowels and char not in count:
        count+=char
print(count)
```
