```python
#count uppercase and lowercase letter in a string
x="Hellow World"
uppercase=""
lowercase=""
for char in x:
    if char.isupper():
        uppercase+=char
    if char.islower():
        lowercase+=char
print("uppercase:",uppercase)
print("lowercase:",lowercase)
```

```python
#check if a string is a valid address
import re
x="test@example.com"
pattern=r'^[\w.-]+@[\w\.-]+\.\w+$'
if re.match(pattern,x):
    print("valid email address")
else:
    print("invalid email address") 
```
